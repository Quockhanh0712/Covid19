```python

# Đảm bảo cột 'date' ở định dạng datetime
vietnam_data['date'] = pd.to_datetime(vietnam_data['date'])

# Biểu đồ số ca nhiễm hàng ngày
plt.figure(figsize=(14, 6))
plt.plot(vietnam_data['date'], vietnam_data['new_cases'], label="New Cases", color='skyblue')
plt.title("Daily New COVID-19 Cases in Vietnam")
plt.xlabel("Date")
plt.ylabel("New Cases")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Biểu đồ tích lũy số ca nhiễm
vietnam_data['cumulative_cases'] = vietnam_data['new_cases'].cumsum()
plt.figure(figsize=(14, 6))
plt.plot(vietnam_data['date'], vietnam_data['cumulative_cases'], label="Cumulative Cases", color='orange')
plt.title("Cumulative COVID-19 Cases in Vietnam")
plt.xlabel("Date")
plt.ylabel("Cumulative Cases")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Đỉnh dịch: Xác định ngày có số ca nhiễm cao nhất
peak_cases = vietnam_data[vietnam_data['new_cases'] == vietnam_data['new_cases'].max()]
print("Ngày có số ca nhiễm cao nhất:")
print(peak_cases[['date', 'new_cases']])
```

```python
# Tính tỷ lệ tử vong và biểu đồ số ca tử vong hàng ngày
vietnam_data['mortality_rate'] = (vietnam_data['total_deaths'] / vietnam_data['total_cases']) * 100

# Biểu đồ số ca tử vong hàng ngày
plt.figure(figsize=(14, 6))
plt.plot(vietnam_data['date'], vietnam_data['new_deaths'], label="New Deaths", color='red')
plt.title("Daily COVID-19 Deaths in Vietnam")
plt.xlabel("Date")
plt.ylabel("New Deaths")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Biểu đồ tỷ lệ tử vong theo thời gian
plt.figure(figsize=(14, 6))
plt.plot(vietnam_data['date'], vietnam_data['mortality_rate'], label="Mortality Rate (%)", color='purple')
plt.title("COVID-19 Mortality Rate in Vietnam Over Time")
plt.xlabel("Date")
plt.ylabel("Mortality Rate (%)")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# So sánh tỷ lệ tử vong với khu vực quốc tế có thể làm như sau:
# Lọc dữ liệu của các nước để so sánh với Việt Nam (thực hiện nếu có dữ liệu đa quốc gia)
```

```python


# Giả sử vietnam_data đã được định nghĩa từ trước

# Lọc dữ liệu từ cuối năm 2021 đến cuối năm 2023
vietnam_data_filtered = vietnam_data[(vietnam_data['date'] >= '2021-04-30') & (vietnam_data['date'] <= '2023-12-31')]

# Tính tỷ lệ dương tính: Số ca nhiễm mới / Số lượng xét nghiệm
vietnam_data_filtered['positivity_rate'] = (vietnam_data_filtered['new_cases'] / vietnam_data_filtered['new_tests']) * 100

# Biểu đồ số lượng xét nghiệm hàng ngày
plt.figure(figsize=(14, 6))
plt.plot(vietnam_data_filtered['date'], vietnam_data_filtered['new_tests'], label="New Tests", color='green')
plt.title("Daily COVID-19 Tests in Vietnam (End of 2021 - End of 2023)")
plt.xlabel("Date")
plt.ylabel("New Tests")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Biểu đồ tỷ lệ dương tính
plt.figure(figsize=(14, 6))
plt.plot(vietnam_data_filtered['date'], vietnam_data_filtered['positivity_rate'], label="Positivity Rate (%)", color='darkblue')
plt.title("COVID-19 Positivity Rate in Vietnam (End of 2021 - End of 2023)")
plt.xlabel("Date")
plt.ylabel("Positivity Rate (%)")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python




# Tính tỷ lệ tiêm chủng (tiêm chủng tích lũy / tổng dân số)
vietnam_data['vaccination_rate'] = vietnam_data['total_vaccinations'] / vietnam_data['population'] * 100

# Vẽ biểu đồ tỷ lệ tiêm chủng theo thời gian
plt.figure(figsize=(14, 6))
plt.plot(vietnam_data['date'], vietnam_data['vaccination_rate'], label="Vaccination Rate (%)", color='green')
plt.title("Vaccination Rate in Vietnam Over Time")
plt.xlabel("Date")
plt.ylabel("Vaccination Rate (%)")
plt.legend()
plt.show()
```

```python
# Vẽ biểu đồ tiêm chủng và ca nhiễm mới theo thời gian
plt.figure(figsize=(14, 6))
plt.plot(vietnam_data['date'], vietnam_data['vaccination_rate'], label="Vaccination Rate (%)", color='green')
plt.plot(vietnam_data['date'], vietnam_data['new_cases'], label="New Cases", color='red')
plt.title("Effect of Vaccination on New Cases")
plt.xlabel("Date")
plt.ylabel("Rate / New Cases")
plt.legend()
plt.show()
```

```python
# Tính tỷ lệ hồi phục (giả sử total_cases - total_deaths = số ca hồi phục)
vietnam_data['recovery_rate'] = (vietnam_data['total_cases'] - vietnam_data['total_deaths']) / vietnam_data['total_cases'] * 100

# Vẽ biểu đồ tỷ lệ hồi phục theo thời gian
plt.figure(figsize=(14, 6))
plt.plot(vietnam_data['date'], vietnam_data['recovery_rate'], label="Recovery Rate (%)", color='blue')
plt.title("Recovery Rate in Vietnam Over Time")
plt.xlabel("Date")
plt.ylabel("Recovery Rate (%)")
plt.legend()
plt.show()
```

```python
# Các khoảng thời gian giãn cách xã hội (ví dụ)
lockdown_periods = [('2020-03-01', '2020-04-30'), ('2021-05-01', '2021-08-31')]

# Vẽ biểu đồ số ca nhiễm và tử vong trong các đợt giãn cách
plt.figure(figsize=(14, 6))
plt.plot(vietnam_data['date'], vietnam_data['new_cases'], label="New Cases", color='red')
plt.plot(vietnam_data['date'], vietnam_data['new_deaths'], label="New Deaths", color='black')

for start, end in lockdown_periods:
    plt.axvspan(pd.to_datetime(start), pd.to_datetime(end), color='gray', alpha=0.3, label="Lockdown Period")

plt.title("COVID-19 Cases and Deaths During Lockdown Periods in Vietnam")
plt.xlabel("Date")
plt.ylabel("New Cases / Deaths")
plt.legend()
plt.show()
```

```python
# Lọc dữ liệu cho biến chủng Delta (từ tháng 5/2021 đến tháng 12/2021)

delta_wave = vietnam_data[(vietnam_data['date'] >= '2021-05-01') & (vietnam_data['date'] <= '2021-12-31')]

# Lọc dữ liệu cho biến chủng Omicron (từ tháng 12/2021 đến giữa năm 2022)
omicron_wave = vietnam_data[(vietnam_data['date'] >= '2021-12-01') & (vietnam_data['date'] <= '2022-06-30')]
```

```python
plt.figure(figsize=(12, 6))
plt.plot(delta_wave['date'], delta_wave['new_cases'], label='Delta Wave', color='blue')
plt.plot(omicron_wave['date'], omicron_wave['new_cases'], label='Omicron Wave', color='orange')
plt.title('So sánh số ca nhiễm COVID-19 giữa biến chủng Delta và Omicron')
plt.xlabel('Ngày')
plt.ylabel('Số ca nhiễm mới')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(delta_wave['date'], delta_wave['new_deaths'], label='Delta Wave', color='red')
plt.plot(omicron_wave['date'], omicron_wave['new_deaths'], label='Omicron Wave', color='green')
plt.title('So sánh số ca tử vong COVID-19 giữa biến chủng Delta và Omicron')
plt.xlabel('Ngày')
plt.ylabel('Số ca tử vong mới')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Ví dụ phân chia thành các giai đoạn
giai_doan_1 = vietnam_data[(vietnam_data['date'] >= '2020-01-01') & (vietnam_data['date'] <= '2020-12-31')]
giai_doan_2 = vietnam_data[(vietnam_data['date'] >= '2021-01-01') & (vietnam_data['date'] <= '2021-06-30')]
giai_doan_3 = vietnam_data[(vietnam_data['date'] >= '2021-07-01') & (vietnam_data['date'] <= '2021-12-31')]
giai_doan_4 = vietnam_data[(vietnam_data['date'] >= '2022-01-01')]
```

```python
# Tính tổng và trung bình cho mỗi giai đoạn
def summarize(data):
    return {
        'Total Cases': data['total_cases'].max(),
        'Total Deaths': data['total_deaths'].max(),
        'Average New Cases': data['new_cases'].mean(),
        'Average New Deaths': data['new_deaths'].mean(),
        'Average Vaccinations': data['new_vaccinations_smoothed'].mean(),
    }

giai_doan_1_summary = summarize(giai_doan_1)
giai_doan_2_summary = summarize(giai_doan_2)
giai_doan_3_summary = summarize(giai_doan_3)
giai_doan_4_summary = summarize(giai_doan_4)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.plot(giai_doan_1['date'], giai_doan_1['new_cases'], label='Giai đoạn 1')
plt.plot(giai_doan_2['date'], giai_doan_2['new_cases'], label='Giai đoạn 2')
plt.plot(giai_doan_3['date'], giai_doan_3['new_cases'], label='Giai đoạn 3')
plt.plot(giai_doan_4['date'], giai_doan_4['new_cases'], label='Giai đoạn 4')
plt.xlabel('Date')
plt.ylabel('New Cases')
plt.title('New COVID-19 Cases in Vietnam by Phase')
plt.legend()
plt.show()
```
